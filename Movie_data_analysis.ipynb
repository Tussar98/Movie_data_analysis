# Handling extreme values

```python
# Function to identify outliers using the IQR method
def identify_outliers_iqr(df, columns):
    # Dictionary to store outliers for each column
    outliers = {}
    bounds = {}  # Dictionary to store bounds for each column for capping later
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Store outliers and bounds for each column
        outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        bounds[column] = (lower_bound, upper_bound)
    return outliers, bounds

# Function to remove outliers from DataFrame
def remove_outliers(df, bounds):
    df_no_outliers = df.copy()
    for column, (lower_bound, upper_bound) in bounds.items():
        df_no_outliers = df_no_outliers[(df_no_outliers[column] >= lower_bound) & (df_no_outliers[column] <= upper_bound)]
    return df_no_outliers

# Function to cap outliers in the DataFrame
def cap_outliers(df, bounds):
    df_capped = df.copy()
    for column, (lower_bound, upper_bound) in bounds.items():
        df_capped[column] = np.where(df_capped[column] < lower_bound, lower_bound, df_capped[column])
        df_capped[column] = np.where(df_capped[column] > upper_bound, upper_bound, df_capped[column])
    return df_capped
```

```python
# Specifing the columns to check for outliers
columns_to_check = ["popularity", "vote_average", "runtime", "vote_count", "budget", "release_year", 
                    "AverageRating", "IMDB_Rating", "Meta_score", "overview_sentiment"]

# Identifing outliers and bounds
outliers, bounds = identify_outliers_iqr(df, columns_to_check)

# Removing outliers
df_no_outliers = remove_outliers(df, bounds)
print("\nDataFrame after removing outliers:")
print(df_no_outliers)

# Caping outliers
df_capped = cap_outliers(df, bounds)
print("\nDataFrame after capping outliers:")
print(df_capped)

# Saving the cleaned DataFrames to new CSV files
df_no_outliers.to_csv('no_outliers_file.csv', index=False)
df_capped.to_csv('capped_outliers_file.csv', index=False)
```

```python
# Remove columns with all values missing to reduce unnecessary columns
# This keeps the DataFrame compact and focused on columns with meaningful data
df = df.dropna(axis=1, how='all')

# Display the updated structure of the DataFrame to understand the remaining columns
# This helps verify that only columns with at least some data were retained
df.info()

# Remove any columns with 'Unnamed' in their name, as these typically result from index issues in CSV files
# Columns named 'Unnamed' are usually empty or irrelevant and can clutter the DataFrame
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Loop through each column in the DataFrame to handle missing values
# Numeric columns will have missing values filled with the mean of the column
# Categorical columns will have missing values filled with the most common value (mode)
for col in df.columns:
    # Check if the column has a numeric data type
    if pd.api.types.is_numeric_dtype(df[col]):
        # Calculate the mean of the numeric column to use for imputation
        mean_value = df[col].mean()
        # Replace missing values in the column with the calculated mean
        df[col] = df[col].fillna(mean_value)  # Assign back the filled column
        print(f"Imputed missing values in numerical column '{col}' with mean: {mean_value}")
    else:
        # For categorical columns, calculate the mode (most frequent value) for imputation
        mode_value = df[col].mode()[0]  # [0] to select the most frequent value
        # Replace missing values in the column with the calculated mode
        df[col] = df[col].fillna(mode_value)  # Assign back the filled column
        print(f"Imputed missing values in categorical column '{col}' with mode: {mode_value}")

# After handling missing values, check if there are any remaining missing values in the DataFrame
# This step helps verify that all columns are now complete with no missing entries
print("\nRemaining missing values per column after imputation:")
print(df.isnull().sum())

# Summary:
# - Columns with all missing values were removed entirely.
# - Columns with 'Unnamed' in their names were dropped, as they likely contain no useful data.
# - Missing values in numeric columns were filled with the column's mean.
# - Missing values in categorical columns were filled with the column's mode.
# This process ensures that the DataFrame has no missing values, making it ready for analysis or modeling.
```
